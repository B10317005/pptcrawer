```python
from bs4 import BeautifulSoup
import requests

class PttCrawer:
    def __init__(self,url,page):
        self.url=url
        self.page=page
        self.session=requests.Session()
        self.session.cookies.update({'over18':'1'})
    def run(self):
        for i in range(0,self.page):
            resp=self.session.get(self.url)
            soup =BeautifulSoup(resp.text.encode("utf-8"),"lxml")
            articles = soup.find_all('div', 'r-ent')
            for article in articles:
                meta = article.find('div', 'title').find('a')
                if meta!=None:
                    title = meta.getText().strip()
                    push = article.find('div', 'nrec').getText()
                    date = article.find('div', 'date').getText()
                    author = article.find('div', 'author').getText()
                    print(title,push,date,author,"\n")
        
                gossip_list=soup.find('div',{'class':'btn-group btn-group-paging'})
        
        gg=gossip_list.find_all('a')[1]['href']
        self.url="https://www.ptt.cc"+gg
```

```python
pttcrawer = PttCrawer('https://www.ptt.cc/bbs/Gossiping/index.html',5)
pttcrawer.run()
```
